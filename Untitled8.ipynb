**Аналіз Часових Рядів та Прогнозування – Прогнозування Продажів для РітейлМережі**

• Мета: Прогнозування майбутніх продажів для рітейл-мережі на основі
історичних даних за допомогою аналізу часових рядів.

• Набір даних: Використовуйте Retail Sales Time Series Dataset (доступний на
Kaggle) або інший набір даних із часовими рядами, що стосуються продажів.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

**1. Завантаження та Огляд Даних:**

Завантажте набір даних із продажами та перегляньте структуру
часових рядів (наприклад, щомісячні або щотижневі дані).

```python
df = pd.read_csv('train.csv')
print(df.head())

print(df.info())
```

Проведіть декомпозицію часових рядів для аналізу тренду,
сезонності та залишків.

```python
#Робимо з тексту дати
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df.set_index('Order Date', inplace=True)

#Рахуємо суму продажів за кожен місяць (M = Month)
monthly_sales = df['Sales'].resample('M').sum()

#Робимо декомпозицію
result = seasonal_decompose(monthly_sales, model='additive')

#Малюємо результат
result.plot()
plt.show()
```

**2. Попередня Обробка Даних:**

Обробіть відсутні дані, видаліть аномалії, перевірте
стаціонарність ряду (за потреби).

```python
# 1. Обробка відсутніх даних
# Якщо є пропуски, заповнюємо їх інтерполяцією
monthly_sales = monthly_sales.interpolate()

# 2. Видалення аномалій
# Використаємо правило "3 сигми": все, що вище/нижче певної межі - аномалія.
mean = monthly_sales.mean()
std = monthly_sales.std()
upper_limit = mean + (3 * std)
lower_limit = mean - (3 * std)
```

Розділіть дані на тренувальний і тестовий набори для
прогнозування.

```python
# Беремо 80% даних на навчання, 20% на тест
train_size = int(len(monthly_sales) * 0.8)

train = monthly_sales.iloc[:train_size]
test = monthly_sales.iloc[train_size:]

print(f"\nРозмір тренувального набору: {len(train)} місяців")
print(f"Розмір тестового набору: {len(test)} місяців")

# Візуалізація розділення
plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Train (Навчання)', color='blue')
plt.plot(test.index, test, label='Test (Перевірка)', color='orange')
plt.title('Розділення даних на тренувальну та тестову вибірки')
plt.legend()
plt.show()
```

**3. Створення Моделі Прогнозування Часових Рядів:**

Реалізуйте модель ARIMA або SARIMA за допомогою statsmodels.

```python
#Використовуємо SARIMA.
#order=(1, 1, 1) - параметри для тренду
#seasonal_order=(1, 1, 1, 12) - параметри для сезонності (12 означає 12 місяців у році)
model = SARIMAX(train,
                order=(1, 1, 1),
                seasonal_order=(1, 1, 1, 12),
                enforce_stationarity=False,
                enforce_invertibility=False)
```

Проведіть навчання моделі та зробіть прогноз на майбутні
періоди.

```python
model_fit = model.fit(disp=False) # disp=False прибирає зайвий текст при навчанні

#Робимо прогноз на довжину тестового набору
start = len(train)
end = len(train) + len(test) - 1
predictions = model_fit.predict(start=start, end=end, typ='levels')
```

Оцініть точність прогнозу за допомогою MAE або RMSE.

```python
mae = mean_absolute_error(test, predictions)
rmse = np.sqrt(mean_squared_error(test, predictions))

print(f"\n--- Результати точності ---")
print(f"MAE (Середня помилка): {mae:.2f}")
print(f"RMSE (Квадратична помилка): {rmse:.2f}")
print(f"Середній обсяг продажів у тесті: {test.mean():.2f}")
print(f"Помилка у відсотках (приблизно): {(mae / test.mean()) * 100:.1f}%")
```

**4. Візуалізація та Презентація Прогнозів:**

Візуалізуйте прогнозовані продажі з інтервалами довіри.

```python
forecast = model_fit.get_forecast(steps=len(test))
predictions = forecast.predicted_mean
conf_int = forecast.conf_int()

plt.figure(figsize=(12, 6))
plt.plot(train, label='Навчальні дані (Train)', color='blue')
plt.plot(test, label='Реальні дані (Test)', color='green')
plt.plot(predictions, label='Прогноз SARIMA', color='red', linestyle='--')
plt.fill_between(conf_int.index,
                 conf_int.iloc[:, 0], # Нижня межа
                 conf_int.iloc[:, 1], # Верхня межа
                 color='pink',        # Колір
                 alpha=0.3,           # Прозорість
                 )

plt.title('Прогноз продажів: Реальність проти Моделі')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
